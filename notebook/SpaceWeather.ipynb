```python
# Updated feature selection
features = icmecat[['mo_sc_heliodistance', 'mo_sc_long_heeq', 'mo_sc_lat_heeq',
               'icme_duration', 'icme_bmax', 'icme_bmean', 'icme_bstd',
               'icme_speed_mean', 'icme_speed_std', 'mo_duration',
               'mo_bmax', 'mo_bmean', 'mo_bstd',
               'mo_speed_mean', 'mo_speed_std',
               'mo_expansion_speed', 'mo_pdyn_mean', 'mo_pdyn_std',
               'mo_density_mean', 'mo_density_std',
               'mo_temperature_mean', 'mo_temperature_std',
               'sheath_speed_mean', 'sheath_speed_std',
               'sheath_density_mean', 'sheath_density_std',
               'sheath_pdyn_mean', 'sheath_pdyn_std']]

# Target variables remain the same
target = icmecat[['mo_bzmean', 'mo_bzmin', 'mo_bzstd']]
```

```python
# Check for missing values
print(features.isnull().sum())
print(target.isnull().sum())
```

```python
!pip install scikit-learn
import numpy as np
from sklearn.model_selection import train_test_split

# Create a copy of the dataset
df_copy = features.copy()

# Randomly mask 10% of the data to simulate missing values
np.random.seed(42)
df_missing = df_copy.mask(np.random.rand(*df_copy.shape) < 0.1)

# Split into train and test sets
df_train, df_test = train_test_split(df_missing, test_size=0.2, random_state=42)
```

```python
df_mean_imputed = df_train.fillna(df_train.mean())
df_median_imputed = df_train.fillna(df_train.median())
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
df_knn_imputed = imputer.fit_transform(df_train)
df_knn_imputed = pd.DataFrame(df_knn_imputed, columns=df_train.columns)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

df_reg_imputed = df_train.copy()

for column in df_train.columns:
    missing_mask = df_train[column].isna()
    if missing_mask.sum() > 0:
        X_train = df_train.loc[~missing_mask].drop(columns=[column])
        y_train = df_train.loc[~missing_mask, column]
        X_missing = df_train.loc[missing_mask].drop(columns=[column])

        # Create a pipeline with imputer and linear regression
        pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('regressor', LinearRegression())
        ])

        pipeline.fit(X_train, y_train)
        df_reg_imputed.loc[missing_mask, column] = pipeline.predict(X_missing)
```

```python

# Reset indices to ensure consistency
df_test = df_test.reset_index(drop=True)
df_mean_imputed = df_mean_imputed.reset_index(drop=True)
df_median_imputed = df_median_imputed.reset_index(drop=True)
df_knn_imputed = df_knn_imputed.reset_index(drop=True)
df_reg_imputed = df_reg_imputed.reset_index(drop=True)
```

```python
!pip install numpy
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def evaluate_imputation(original, imputed, method_name):
    mae = mean_absolute_error(original, imputed)
    mse = mean_squared_error(original, imputed)
    rmse = np.sqrt(mse)
    print(f"{method_name} - MAE: {mae}, MSE: {mse}, RMSE: {rmse}")

# Iterate over columns and evaluate
for column in df_test.columns:
    original = df_test[column].dropna()

    # Impute only the columns that have missing values
    if original.index.isin(df_mean_imputed.index).all():
        imputed_mean = df_mean_imputed.loc[original.index, column]
        imputed_median = df_median_imputed.loc[original.index, column]
        imputed_knn = df_knn_imputed.loc[original.index, column]
        imputed_reg = df_reg_imputed.loc[original.index, column]

        evaluate_imputation(original, imputed_mean, "Mean Imputation")
        evaluate_imputation(original, imputed_median, "Median Imputation")
        evaluate_imputation(original, imputed_knn, "k-NN Imputation")
        evaluate_imputation(original, imputed_reg, "Regression Imputation")
```

Median Imputation seems to perform the best overall based on the provided metrics. It generally has the lowest MAE, MSE, and RMSE, suggesting it provides the most accurate imputation with the least error.

```python
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_f=correlation(features,0.85)
```

```python
## drop features when correlation is more than 0.85
features.drop(corr_f,axis=1,inplace=True)
```

```python
## getting all different of features
num_features= [feature for feature in features.columns if features[feature].dtype!='O']
print('Num of numerical features:',len(num_features))
cat_features=[feature for feature in features.columns if features[feature].dtype=='O']
print('Num of categorical features:',len(cat_features))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

# Assuming 'num_features' is a list of numerical column names from your previous code
numeric_transformer = StandardScaler()

preprocessor = ColumnTransformer(
    [
        ('StandardScaler', numeric_transformer, num_features),  # Apply StandardScaler only to numerical columns
    ],
    remainder='passthrough'  # Keep the rest of the columns unchanged
)

features_scaled = preprocessor.fit_transform(features)
```

```python
## separate train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=42)
x_train.shape,x_test.shape
```

```python
!pip install xgboost
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, MultiTaskLassoCV, ElasticNet, MultiTaskElasticNetCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
## create a function to evalute model
def evalute_model(true, predicted):
    mae=mean_absolute_error(true,predicted)
    mse=mean_squared_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_square=r2_score(true,predicted)
    return mae, mse, rmse, r2_square
```

```python
## beginning the model training

models = {
    'Linear Regression': MultiOutputRegressor(LinearRegression()),  # Wrap in MultiOutputRegressor
    'MultiTaskLassoCV': MultiTaskLassoCV(cv=5),  # Already supports multi-output
    'Ridge': MultiOutputRegressor(Ridge()),  # Wrap in MultiOutputRegressor
    'RidgeCV': MultiOutputRegressor(RidgeCV()),  # Wrap in MultiOutputRegressor
    'ElasticNet': MultiOutputRegressor(ElasticNet()),  # Wrap in MultiOutputRegressor
    'ElasticNetCV': MultiTaskElasticNetCV(),  # Already supports multi-output
    'K-Neighbors Regressor': MultiOutputRegressor(KNeighborsRegressor()),  # Wrap in MultiOutputRegressor
    'Decision Tree': MultiOutputRegressor(DecisionTreeRegressor()),  # Wrap in MultiOutputRegressor
    "Random Forest": MultiOutputRegressor(RandomForestRegressor()),  # Wrap in MultiOutputRegressor
    "Xgboost": MultiOutputRegressor(XGBRegressor()),  # Wrap in MultiOutputRegressor
    'Gradient Boosting': MultiOutputRegressor(GradientBoostingRegressor()),  # Wrap in MultiOutputRegressor
    'AdaBoost': MultiOutputRegressor(AdaBoostRegressor())  # Wrap in MultiOutputRegressor
}


for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)

    # make predictions
    y_train_pred= model.predict(x_train)
    y_test_pred= model.predict(x_test)

    model_train_mae,model_train_mse,model_train_rmse,model_train_r2=evalute_model(y_train,y_train_pred)
    model_test_mae,model_test_mse,model_test_rmse,model_test_r2=evalute_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print('Model performance for Training set')
    print('Root Mean Squared Error:',model_train_rmse)
    print('Mean Absolute Error:',model_train_mae)
    print('R2 Score:',model_train_r2)
    print()
    print('Model performance for Test set')
    print('Root Mean Squared Error:',model_test_rmse)
    print('Mean Absolute Error:',model_test_mae)
    print('R2 Score:',model_test_r2)
    print('='*30)
```

Linear Regression, ElasticNet and Gradient Boosting are working with great efficiency.

```python
## Working on different parameters for these models
LR_params = {'alpha': [0.1, 1, 10, 100]}
GB_params = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 500],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 1.0],
    'max_features': ['auto', 'sqrt', 'log2']
}
EN_params = {
    'alpha': [0.1, 1, 10],
    'l1_ratio': [0.1, 0.5, 0.9]
}
```

```python
## model list for hyperparameter tuning
models=[
    ('Ridge',MultiOutputRegressor(Ridge()),LR_params),
    ('EN',MultiOutputRegressor(ElasticNet()),EN_params),
    ('GB',MultiOutputRegressor(GradientBoostingRegressor()),GB_params)
]
```

```python
## Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV

model_param={}
for name, model, params in models:
    # Access the parameters of the inner estimator
    param_grid = {f'estimator__{key}': value for key, value in params.items()}

    random=GridSearchCV(estimator=model,
                             param_grid=param_grid,  # Use the modified param_grid
                             cv=5,
                             verbose=3,
                             n_jobs=-1,
                            scoring='neg_mean_squared_error')
    random.fit(x_train,y_train)
    model_param[name]=random.best_params_

for model_name in model_param:
    print(f"Best Params for {model_name}")
    print(model_param[model_name])
```

```python
models={
    'Ridge':MultiOutputRegressor(Ridge(alpha=0.1)),
    'ElasticNet':MultiOutputRegressor(ElasticNet(alpha=10, l1_ratio=0.9)),
    'GradientBoostingRegressor':MultiOutputRegressor(GradientBoostingRegressor(learning_rate=0.2, max_depth=3, max_features= 'log2', min_samples_leaf=1, n_estimators=100, min_samples_split=2, subsample=1.0))
}

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    
    # make predictions
    y_train_pred= model.predict(x_train)
    y_test_pred= model.predict(x_test)
    
    model_train_mae,model_train_mse,model_train_rmse,model_train_r2=evalute_model(y_train,y_train_pred)
    model_test_mae,model_test_mse,model_test_rmse,model_test_r2=evalute_model(y_test,y_test_pred)
    
    print(list(models.keys())[i])
    print('Model performance for Training set')
    print('Root Mean Squared Error:',model_train_rmse)
    print('Mean Absolute Error:',model_train_mae)
    print('R2 Score:',model_train_r2)
    print()
    print('Root Mean Squared Error:',model_test_rmse)
    print('Mean Absolute Error:',model_test_mae)
    print('R2 Score:',model_test_r2)
    print('='*30)
```
